## Building the Model

```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
emp_df_dummies.head()
```

```python
X = emp_df_dummies.drop(columns = 'Attrition')
Y = emp_df_dummies['Attrition']
```

```python
print(X.shape)
print(Y.shape)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size = 0.2 ,random_state = 212)
```

```python
print("X_Train - ",x_train.shape)
print("X_Test - ",x_test.shape)
print("Y_Train - ",y_train.shape)
print("Y_Test - ",y_test.shape)
```

```python
model_dt = DecisionTreeClassifier(criterion = 'gini' , max_depth = 8, min_samples_leaf = 8 , random_state = 343)
```

```python
model_dt.fit(x_train ,y_train)
```

```python
y_results = model_dt.predict(x_test)
```

Lets Check the model's accuracy score :

```python
model_dt.score(x_test , y_test)
```

Confusion Matrix

```python
print(confusion_matrix(y_test , y_results))
```

```python
# We can see a large in balance in True Positives and True Negatives
```

Classification Report :

```python
print(classification_report(y_test, y_results))
```

```python
# Here we can see the precision , recall , f1-score are very less .
```

```python
# This can be explained :

emp_df['Attrition'].value_counts()
```

#### We can clearly see there is a case of **class imbalancing** where the number of churned customers are far less than the customers who are not churning

```python
# Thus here we will use SMOTEENN :
sm = SMOTEENN()
x_resampled , y_resampled = sm.fit_resample(X,Y)
```

```python
x_train_r , x_test_r , y_train_r , y_test_r = train_test_split(x_resampled,y_resampled,test_size = 0.2 ,random_state = 212)
```

```python
model_dt_SMOTE = DecisionTreeClassifier(criterion = 'gini' , max_depth = 8, min_samples_leaf = 8 , random_state = 343)
```

```python
model_dt_SMOTE.fit(x_train_r , y_train_r)
```

```python
y_predicted = model_dt_SMOTE.predict(x_test_r)
```

Model's Accuracy

```python
model_dt_SMOTE.score(x_test_r,y_test_r)
```

Confusion Matrix :

```python
print(confusion_matrix(y_test_r , y_predicted))
```

Classification Report :

```python
print(classification_report(y_test_r , y_predicted))
```

```python
# Here we can see that our precision , f1_score and recall has all improved .
```

```python
rf_smote.fit(x_train_r , y_train_r)
```

```python
y_new_predicted = rf_smote.predict(x_test_r)
```

```python
rf_smote.score(x_test_r , y_test_r)
```

```python
print(confusion_matrix(y_test_r , y_new_predicted))
```

```python
print("Classification Report is: \n" , classification_report(y_test_r , y_new_predicted))
```
